# Concatination, Merge, Join function

```python
'''
.concat() simply stacks multiple DataFrame together either vertically, or stitches horizontally after aligning on index
.merge() first aligns two DataFrame' selected common column(s) or index, and then pick up the remaining columns from the aligned rows of each DataFrame.

More specifically, .concat():-
Is a top-level pandas function
Combines two or more pandas DataFrame vertically or horizontally
Aligns only on the index when combining horizontally
Errors when any of the DataFrame contains a duplicate index.
Defaults to outer join with the option for inner join

And .merge():-
Exists both as a top-level pandas function and a DataFrame method (as of pandas 1.0)
Combines exactly two DataFrame horizontally
Aligns the calling DataFrame's column(s) or index with the other DataFrame's column(s) or index
Handles duplicate values on the joining columns or index by performing a cartesian product
Defaults to inner join with options for left, outer, and right
Note that when performing pd.merge(left, right), if left has two rows containing the same values from the joining columns or index, 
each row will combine with right's corresponding row(s) resulting in a cartesian product. 
On the other hand, if .concat() is used to combine columns, we need to make sure no duplicated index exists in either DataFrame.

Practically speaking:

Consider .concat() first when combining homogeneous DataFrame, while consider .merge() first when combining complementary DataFrame.
If need to merge vertically, go with .concat(). If need to merge horizontally via columns, go with .merge(), which by default merge on the columns in common.
'''
```

```python
import pandas as pd

dict1 = pd.DataFrame({
    'a1':['a1','a2','a3','a4'],
    'b1':['b1','b2','b3','b4'],
    'c1':['c1','c2','c3','c4'],
    'd1':['d1','d2','d3','d4'],
},index= [1,2,3,4])

dict2 = pd.DataFrame({
    'a2':['a5','a6','a7','a8'],
    'b2':['b5','b6','b7','b8'],
    'c2':['c5','c6','c7','c8'],
    'd2':['d5','d6','d7','d8'],
}, index=[5,6,7,8])

dict3 = pd.DataFrame({
    'a3':['a9','a10','a11','a12'],
    'b3':['b9','b10','b11','b12'],
    'c3':['c9','c10','c11','c12'],
    'd3':['d9','d10','d11','d12'],
}, index=[9,10,11,12])

pd.concat([dict1,dict2,dict3], axis=0) # since column name are different so it arraged in rows as well in column (this is not i wanted)
```

## Object oriented method

```python
# In object-oriented interface, Pyplot is used only for a few functions such as figure creation, 
# and the user explicitly creates and keeps track of the figure and axes objects. 
# At this level, the user uses Pyplot to create figures, and through those figures, 
# one or more axes objects can be created. These axes objects are then used for most plotting actions.
```

```python
# step-1:- To begin with, we create a figure instance which provides an empty canvas.
fig = plt.figure()

# step-2:- Now add axes to figure. 
# The add_axes() method requires a list object of 4 elements corresponding to left, bottom, width and height of the figure. 
# Each number must be between 0 and 1.
ax=fig.add_axes([0,0,1,1])

# step-3:- plotting of x and y
ax.plot(x,y)

# step-3:- Set labels for x and y axis as well as title.
ax.set_title("x vs y")
ax.set_xlabel('value of x')
ax.set_ylabel('value of y')

plt.show()
```

```python
fig = plt.figure()

ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.2,0.5,0.5])

ax1.plot(x,y)
ax2.plot(x,y)

ax1.set_title("x vs y 1")
ax1.set_xlabel('value of x')
ax1.set_ylabel('value of y')

ax2.set_title("x vs y 2")
ax2.set_xlabel('value of x')
ax2.set_ylabel('value of y')

plt.show()
```

# Subplot Method

```python
fig,axes = plt.subplots(nrows=1,ncols=2, figsize=(2,2))
axes[0].plot(x,y)
axes[1].plot(x,y)
axes[0].set_title('Plot 1')
axes[1].set_title('Plot 2')
```

# Figure size, Aspect ratio and DPI

```python
axis_font = {'fontname':'Arial', 'size':'14'}
title_font = {'fontname':'Arial', 'size':'16', 'color':'black', 'weight':'normal', 'verticalalignment':'bottom'}

fig,axes = plt.subplots(nrows=2, ncols=2, figsize=(5,5), dpi=100) # dpi means dots per inch, more dots more clearity

axes[0,0].plot(x,y, color='r', marker='*', linestyle='--', linewidth=1.5, markersize=5)
axes[0,1].plot(x,y)
axes[1,0].plot(x,y)
axes[1,1].plot(x,y)
axes[0,0].set_title('Plot 1', **title_font)
axes[0,1].set_title('Plot 2')
axes[1,0].set_title('Plot 1')
axes[1,1].set_title('Plot 2')

plt.tight_layout()
```

# Legends

```python
x = [i for i in range(0,11)]
y1 = [i**2 for i in x]
y2 = [i**3 for i in x]
fig,ax = plt.subplots(1,2,figsize=(8,8))
ax[0].plot(x,y1,color='r', marker='*', label='square')  # for legends lable is required in plot function.
ax[0].set_xlabel('value of x')
ax[0].set_ylabel('value of y1')
ax[0].legend(loc=0)

ax[1].plot(x,y2, color='g', marker='o', label='cube')   # for legends lable is required in plot function.
ax[1].set_xlabel('value of x')
ax[1].set_ylabel('value of y2')
ax[1].legend(loc=0)

plt.tight_layout()
```

# x and y lim, decorating part

```python
x = [i for i in range(-10,11)]
y1 = [i**2 for i in x]
y2 = [i**3 for i in x]

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10,10), dpi=100)
axes[0,0].plot(x,y1, color='r', linestyle='--', linewidth=1.5, marker='*', markersize=10, alpha=0.2, markerfacecolor='yellow') # alpha used for to set transperancy of plot
axes[0,1].plot(x,y1, color='r', linestyle='--', linewidth=3.0, marker='*', markersize=10, alpha=0.8, markerfacecolor='yellow') # by default alpha = 1.0
axes[1,0].plot(x,y2, color='m', linestyle='-.', linewidth=1.5, marker='o', markersize=10, alpha=1.0, markerfacecolor='red')
axes[1,1].plot(x,y2, color='m', linestyle=':', linewidth=3.0, marker='+', markersize=3.0, alpha=0.5, markerfacecolor='green')
axes[1,1].grid()
axes[1,1].set_xlim([-10,10])  # see the effet of x limit
#axes[1,1].set_ylim([-20,20])
```

## bar chart

```python
def a(y):                      # for adding value onthe top of the bar plot this function has been defined here
  for i,j in enumerate(y):
    plt.text(i+0.80, y=j+10, s=str(j))
  return plt.text

a(y_views)
plt.bar(days, y_views, color='y', label='youtube views')
plt.bar(days, f_views, color='g', label='facebook views')
#plt.text()                    # plt.text(x-coordinate, y-coordinate, string of <value>) is the function is for adding value on the top of the bar plot
plt.legend(loc=0)
```
